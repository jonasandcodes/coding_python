```python
import pandas as pd
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/jonasandcodes/trabalho-alg.log.pro/main/cambio_historico_diario.csv'

date_format = '%d/%m/%Y'
cambio = pd.read_csv(url, delimiter = ';', header=0, index_col=0, parse_dates=True,
                     date_parser=lambda x: pd.to_datetime(x, format=date_format))

cambio['valor'] = pd.to_numeric(cambio['valor'].str.replace(',', '.'))

data_inicio = pd.to_datetime('2022/02/01')
data_fim = pd.to_datetime('2024/02/01')

cambio.f = cambio.loc[(cambio.index >= data_inicio) & (cambio.index <= data_fim)]

cambio.f
```

```python
url = 'https://raw.githubusercontent.com/jonasandcodes/trabalho-alg.log.pro/main/pre%C3%A7o_petr_bruto_historico_brent.csv'

date_format = '%d/%m/%Y'
petroleo = pd.read_csv(url, delimiter=',', header = 0, index_col=0, parse_dates=True,
                       date_parser=lambda x: pd.to_datetime(x, format=date_format))

data_inicio = pd.to_datetime('2022-02-01')
data_fim = pd.to_datetime('2024-02-01')

petroleo.f = petroleo.loc[(petroleo.index >= data_inicio) & (petroleo.index <= data_fim)]

petroleo.f
```

```python
url = 'https://raw.githubusercontent.com/jonasandcodes/trabalho-alg.log.pro/main/pre%C3%A7o_petr_bruto_historico_wti.csv'

date_format = '%d/%m/%Y'
petroleo2 = pd.read_csv(url, delimiter=',', header = 0, index_col=0, parse_dates=True,
                       date_parser=lambda x: pd.to_datetime(x, format=date_format))

data_inicio = pd.to_datetime('2022-02-01')
data_fim = pd.to_datetime('2024-02-01')

petroleo2.f = petroleo2.loc[(petroleo2.index >= data_inicio) & (petroleo2.index <= data_fim)]

petroleo2.f
```

```python
url = 'https://raw.githubusercontent.com/jonasandcodes/trabalho-alg.log.pro/main/S&P_500_historico_NASDAQ.csv'

date_format = '%m/%d/%Y'
sp500 = pd.read_csv(url, delimiter=',', header = 0, index_col=0, parse_dates=True,
                  date_parser=lambda x: pd.to_datetime(x, format=date_format))

data_inicio = pd.to_datetime('2022-01-02')
data_fim = pd.to_datetime('2024-01-02')

sp500.f = sp500.loc[(sp500.index >= data_inicio) & (sp500.index <= data_fim)]

sp500.f
```

```python
url = 'https://raw.githubusercontent.com/jonasandcodes/trabalho-alg.log.pro/main/ibovespa_historico_diario1.csv'

date_format = '%d/%m/%Y'
ibovespa = pd.read_csv(url, delimiter=',', header = 0, index_col=0, parse_dates=True,
                  date_parser=lambda x: pd.to_datetime(x, format=date_format))

data_inicio = pd.to_datetime('2022-02-01')
data_fim = pd.to_datetime('2024-02-01')

ibovespa.f = ibovespa.loc[(ibovespa.index >= data_inicio) & (ibovespa.index <= data_fim)]

ibovespa.f
```

```python
merged_data1 = pd.merge(ibovespa.f, cambio.f, how='outer', left_index=True, right_index=True)
merged_data2 = pd.merge(merged_data1, petroleo.f, how='outer', left_index=True, right_index=True)
merged_data3 = pd.merge(merged_data2, petroleo2.f, how='outer', left_index=True, right_index=True)
merged_data4 = pd.merge(merged_data3, sp500.f, how='outer', left_index=True, right_index=True)
merged_data4
```

```python
columns_to_drop = ['Unnamed: 2_x', 'Unnamed: 2_y', 'Unnamed: 2', 'Open',
                   'High', 'Low']
new_data = merged_data4.drop(columns_to_drop, axis=1)
new_data
```
